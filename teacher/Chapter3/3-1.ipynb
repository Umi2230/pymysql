```python
import pymysql
from configparser import ConfigParser

config = ConfigParser()
config.read('../Chapter1/config.ini')

connection = pymysql.connect(
    host=config.get('DB', 'host'),
    user=config.get('DB', 'user'),
    password=config.get('DB', 'password'),
    port=config.getint('DB', 'port'),
    cursorclass=pymysql.cursors.DictCursor,
)

# 建立資料庫
with connection.cursor() as cursor:
    sql = """
    CREATE DATABASE IF NOT EXISTS chapter3;
    """
    cursor.execute(sql)

    # 建立有時間紀錄的表格
    # 包含 last_login 欄位 預設在寫入時使用現在的時間
    sql = """
        CREATE TABLE IF NOT EXISTS chapter3.users (
            id INT AUTO_INCREMENT PRIMARY KEY,
            name VARCHAR(255) NOT NULL,
            age INT,
            username VARCHAR(255) NOT NULL UNIQUE,
            password VARCHAR(255) NOT NULL,
            last_login TIMESTAMP DEFAULT CURRENT_TIMESTAMP
        );
    """

    cursor.execute(sql)
    cursor.execute("SHOW TABLES FROM chapter3;")
    result = cursor.fetchall()

print(result)
```

```python
# 比較建議的寫法
from datetime import datetime

# 取得當前時間
now = datetime.now()
print(now)
print(now.year)
print(now.month)
print(now.day)
print(now.date())

print("="*30)
# 取得指定時間的 datetime 物件
date = datetime(2025, 1, 1, 15, 30, 0)
print(date)

print("="*30)
# 輸出成指定格式
date_str = date.strftime("%Y/%m/%d")
print(date_str)

date_str = now.strftime("%Y/%m/%d - %H:%M:%S") # 可以任意組合
print(date_str)

date_str = now.strftime("%Y/%m/%d %H:%M:%S")
print(f"[INFO] {date_str} - 程式開始執行")

print("="*30)
# 取得民國年
now = datetime.now()
print(f"民國 {now.year-1911} 年 {now.month} 月 {now.day} 日")
```

```python
# 將指定格式的字串轉換成 datetime 物件
date_string = "2025/11/20 22:11:15"

# 轉換成 datetime 物件 -> datetime.strptime(<datetime str>, <format>)
datetime_obj = datetime.strptime(date_string, "%Y/%m/%d %H:%M:%S")
print(datetime_obj)
print(datetime_obj.date())
```

```python
# 取得前一天日期
from datetime import datetime, timedelta

yesterday = datetime.now() - timedelta(days=1)
last_year = datetime.now() - timedelta(days=365)
print(yesterday)
print(last_year.year)
```

```python
# ISO format
iso_string = '2025-01-01T12:00:00'

# 轉換成 datetime 物件
datetime_obj = datetime.fromisoformat(iso_string)
print(datetime_obj)
```

```python
import calendar

month_calendar = calendar.month(2026, 7)
print(month_calendar)

days_in_month = calendar.monthrange(2026, 7)
print(days_in_month) # -> tuple
print(days_in_month[0]) # 1號是禮拜幾
print(days_in_month[1]) # 這個月最後一天是幾號
```

```python
# 產出 2025 每個月的日期
"""
{
    1: [2025-01-01, 2025-01-02.....2025-01-31],
    2: [2025-02-01, 2025-02-02.....2025-02-28]
}
"""
from pprint import pprint

result = {}
for i in range(1, 13):
    dates = []
    days_max = calendar.monthrange(2025, i)[1]
    for day in range(1, days_max+1):
        dates.append(datetime(2025, i, day).strftime('%Y-%m-%d'))

    result[i] = dates

for i in range(1, 13):
    print(f"{result[i]}")
```
